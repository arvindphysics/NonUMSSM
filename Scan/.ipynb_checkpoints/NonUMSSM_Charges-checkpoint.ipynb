```python
import pandas as pd
```

```python
# Basis: Qu1, Qu2, Qu3, Qd1, Qd2, Qd3, Qe1, Qe2, Qe3, Qq1, Qq2, Qq3, Ql1, Ql2, Ql3, 
# Qv1, Qv2, Qv3, QHu, QHd, Qs, QDx, QDxbar
ChargeSets = pd.read_csv('/Users/oozdal/projects/MagMoment/Scan/NonUMSSMChargeSetsv4.csv', 
                              header=0,sep=",", engine='python')

ChargeSets = ChargeSets.drop_duplicates()  # Drop duplicates
ChargeSets 
```

```python
# Qu1 = Qu2 = Qu3
# Qd1 = Qd2 = Qd3
# Qq1 = Qq2 = Qq3

# Let's print the following basis for simplicity: 
# Qu1, Qd1, Qe1, Qe2, Qe3, Qq1, Ql1, Ql2, Ql3, Qv1, Qv2, Qv3, QHu, QHd, Qs, QDx, QDxbar

ChargeSets = ChargeSets[['Qu1', 'Qd1','Qe1','Qe2','Qe3','Qq1','Ql1','Ql2','Ql3','Qv1','Qv2','Qv3',
                         'QHu','QHd','Qs', 'QDx', 'QDxbar']]

# Renaming the column names -> Qu1 = Qu, Qd1 = Qd, Qq1= Qq
ChargeSets.columns =['Qu', 'Qd','Qe1','Qe2','Qe3','Qq','Ql1','Ql2','Ql3','Qv1','Qv2','Qv3','QHu','QHd','Qs', 'QDx', 'QDxbar']  

ChargeSets
```

```python
# Let's apply some conditions to reduce the number of sets.
# Let's assume that Qe1/Qe2 > 5 and Ql1/Ql2 > 4 
# and absolute value of the followings Qu > 0, Qd>0, Qe1 > 0, Ql1> 0, Ql2 > 0, 

ChargeSets1 = ChargeSets[  
                         (abs(ChargeSets['Qu'])  > 0)  & (abs(ChargeSets['Qd'])  > 0)  &
                         (abs(ChargeSets['QHu']) > 0)  & (abs(ChargeSets['QHd']) > 0)  &
                         (abs(ChargeSets['Qq'])  > 0)  
                        ]

ChargeSets1
```

```python
ChargeSets2 = ChargeSets[  
                         (abs(ChargeSets['Qu'])  > 0)  & (abs(ChargeSets['Qd'])  > 0)  &
                         (abs(ChargeSets['QHu']) > 0)  & (abs(ChargeSets['QHd']) > 0)  & (abs(ChargeSets['Qq']) > 0) &     
                         ((abs(ChargeSets['Qe1'])/abs(ChargeSets['Qe2'])) > 2.) & ((abs(ChargeSets['Ql1'])/abs(ChargeSets['Ql2']) > 2.)) 
                        ]

ChargeSets2
```

```python
ChargeSets3 = ChargeSets[  
                         (abs(ChargeSets['Qu'])  > 0)  & (abs(ChargeSets['Qd'])  > 0)  &
                         (abs(ChargeSets['QHu']) > 0)  & (abs(ChargeSets['QHd']) > 0)  & (abs(ChargeSets['Qq']) > 0) &     
                         ((abs(ChargeSets['Qe1'])/abs(ChargeSets['Qe2'])) > 2.) & ((abs(ChargeSets['Ql1'])/abs(ChargeSets['Ql2']) > 2.5)) 
                        ]

ChargeSets3        
```

```python
ChargeSets4 = ChargeSets[  
                         (abs(ChargeSets['Qu'])  > 0)  & (abs(ChargeSets['Qd'])  > 0)  &
                         (abs(ChargeSets['QHu']) > 0)  & (abs(ChargeSets['QHd']) > 0)  & (abs(ChargeSets['Qq']) > 0) &     
                         ((abs(ChargeSets['Qe1'])/abs(ChargeSets['Qe2'])) > 2.5) & ((abs(ChargeSets['Ql1'])/abs(ChargeSets['Ql2']) > 2.5)) 
                        ]

ChargeSets4      
```
