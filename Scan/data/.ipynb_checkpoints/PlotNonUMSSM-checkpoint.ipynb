```python
df = pd.read_csv('/Users/oozdal/projects/MagMoment/Scan/data/NonUMSSMv1.csv', header=None,sep=" ")
df.columns =["MZp", "ZpWidthOverMass","BRZptoll","BRZptoqq","alpha","gpSUSY","DAEL","DAMU","Chi1","Chi2","Cha1","BRCha1toWChi1", "BRZptoCha1Cha1", "chi2", "Sv1", "TanBeta", \
           "Lambda","TLambda","vS", "NMIX11","NMIX12","NMIX13","NMIX14","NMIX15","NMIX16", "Se1", "M1", "M2", "M3", "M4", "Tep", "Tmup", "Tdp11", "Tdp22", "Tdp33"]

# Let's see a few lines 
#df.head()
```

```python
# Let's slice and reorder the dataframe
df_sub = df[['DAEL','DAMU','alpha','vS','gpSUSY','MZp','Lambda','TLambda','Tep','Tmup','Tdp11','Tdp22','Tdp33','M1','M2','M3','M4', \
            'TanBeta','Chi1','Cha1']]
# Let's see a few lines 
#df_sub.head()
```

```python
# Let's see the solutions with electron g-2 (DAEL) is negative and muon g-2 (DAMU) is positive!
df_sub_DA = df_sub[(df_sub['DAEL'] < 0) & (df_sub['DAMU'] > 0)]

# Let's see a few lines 
#df_sub_DA.head()
```

```python
# Let's see the solutions where muon g-2 (DAMU) within 1,2,3 sigma!

df_sub_DAMU1sig = df_sub_DA[((df_sub_DA['DAMU'])*10**10 > 20.7) & ((df_sub_DA['DAMU'])*10**10 < 36.7)]
df_sub_DAMU2sig = df_sub_DA[((df_sub_DA['DAMU'])*10**10 > 12.7) & ((df_sub_DA['DAMU'])*10**10 < 44.7)]
df_sub_DAMU3sig = df_sub_DA[((df_sub_DA['DAMU'])*10**10 >  3.4) & ((df_sub_DA['DAMU'])*10**10 < 55.6)]
```
